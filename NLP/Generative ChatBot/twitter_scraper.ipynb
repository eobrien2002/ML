```python
import pandas as pd
df_tweets=pd.read_csv('tweets.csv')
df_replies=pd.read_csv('replies.csv')
```

```python
df_merged=pd.merge(df_tweets,df_replies,on='tweet_ID')
```

```python
df_merged.drop(columns={'Unnamed: 0_x','Unnamed: 0_y'},inplace=True)
```

```python
df_merged.head()
```

```python
import re
output = ''
for i in range(len(df_merged)):
    text_x = df_merged.iloc[i]['Text_x'].replace('\n', '')
    text_y = df_merged.iloc[i]['Text_y'].replace('\n', '')
    
    text_x = re.sub(r'#\w+\s?', '', text_x)
    text_y = re.sub(r'#\w+\s?', '', text_y)
 
    text_x = re.sub(r'https?://[A-Za-z0-9./]+', '', text_x)
    text_y = re.sub(r'https?://[A-Za-z0-9./]+', '', text_y)
    
    #text_x = re.sub(r'@\w+\s?', '', text_x)
    #text_y = re.sub(r'@\w+\s?', '', text_y)

    emoji_pattern = re.compile("["
        u"\U0001F600-\U0001F64F"  # emoticons
        u"\U0001F300-\U0001F5FF"  # symbols & pictographs
        u"\U0001F680-\U0001F6FF"  # transport & map symbols
        u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
        u"\U00002702-\U000027B0"
        u"\U000024C2-\U0001F251"
        "]+", flags=re.UNICODE)

    text_x = emoji_pattern.sub(r'', text_x)
    text_y = emoji_pattern.sub(r'', text_y)


    text_x = text_x.replace('\u200b', '')
    text_y = text_y.replace('\u200b', '')

    text_x = re.sub(r'\u2069', '', text_x)
    text_y = re.sub(r'\u2069', '', text_y)

    text_x = re.sub(r'\u2066', '', text_x)
    text_y = re.sub(r'\u2066', '', text_y)

    text_x=text_x.strip()
    text_y=text_y.strip()
    
    if not text_x or not text_y or text_x=='' or text_y=='':
        continue
    output += text_x + '\n' + text_y + '\n'

# write the string to a text file
with open('textfile.txt', 'w',encoding='utf-8') as f:
    f.write(output)
```
